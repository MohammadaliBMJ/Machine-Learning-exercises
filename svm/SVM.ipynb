## Data Preprocessing
For practice we perform Data Preprocessing in both python and postgresql. We first store the csv file as a table in postgresql.

```python
df = pd.read_csv("diabetes.csv")
connection = create_engine("postgresql://postgres:1234512345@172.20.160.1:5432/diabetes")
df.to_sql("diabetes", con = connection, if_exists = 'replace', index = False)
```
